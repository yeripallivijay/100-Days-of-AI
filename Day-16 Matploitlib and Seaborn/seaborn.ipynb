```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.info()
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(
    x='total_bill',
    y='tip',
    data=tips
    )
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(
    x='size',
    y='total_bill',
    data=tips
    )
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
sns.barplot(
    x='day',
    y='total_bill',
    data=tips
    )
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
sns.boxplot(
    x="day",
    y='total_bill',
    data=tips
    )
```

```python
sns.violinplot(
    x='day',
    y='total_bill',
    data=tips
    )
```

```python
sns.histplot(
    tips['total_bill'],
    bins=10,
    kde=True
    )
```

```python
sns.kdeplot(
    tips['total_bill'],
    fill=True
    )
```

```python
sns.pairplot(tips)
```

```python
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
    )
```
