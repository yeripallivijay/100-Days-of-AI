## Stemming

```python
# It helps to solve the classification problems
# Comments of the product is a positive review or negative review
#Reviews --> eating,eat,eaten

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

#### Types of Stemmer:-
    -PorterStemmer
    -RegexpStemmer
    -Snowball Stemmer


#### PorterStemmer

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()

for word in words:
    print(word+"--->"+stemming.stem(word))
```

```python
stemming.stem("sitting")
```

#### RegexpStemmer

```python
from nltk.stem import RegexpStemmer

reg_stem = RegexpStemmer('ing$|s$|e$|able$', min=4)
reg_stem.stem('eating')
```

```python
reg_stem.stem('ingeating')
```

#### Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("english")

for word in words:
    print(word+"--->"+stemmer.stem(word))
```

```python
stemming.stem("fairly"),
stemming.stem("sportingly")
```

```python
stemmer.stem("fairly"),
stemmer.stem("sportingly")
```
